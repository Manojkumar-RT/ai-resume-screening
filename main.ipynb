```python
import spacy
import pandas as pd
import pdfplumber
import re
import os
from docx import Document

nlp = spacy.load("en_core_web_sm")

print("All libraries loaded successfully")
```

```python
folder_path = "resumes"

files = os.listdir(folder_path)
print(files)
```

```python
def extract_text_from_pdf(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            if page.extract_text():
                text += page.extract_text()
    return text
```

```python
def extract_text_from_docx(path):
    doc = Document(path)
    return "\n".join([para.text for para in doc.paragraphs])
```

```python
def extract_contact_details(text):
    email = re.findall(r'\S+@\S+', text)
    phone = re.findall(r'\b\d{10}\b', text)

    email = email[0] if email else ""
    phone = phone[0] if phone else ""

    return email, phone
```

```python
def extract_name(text):
    doc = nlp(text)

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return ent.text

    return "Not Found"
```

```python
skills_list = [
    "python","java","c++","sql","machine learning","deep learning",
    "data science","tableau","power bi","excel","html","css",
    "javascript","react","nodejs","nlp","tensorflow","pandas"
]

def extract_skills(text):
    text = text.lower()
    found = []

    for skill in skills_list:
        if skill in text:
            found.append(skill)

    return found
```

```python
def extract_experience(text):
    exp = re.findall(r'(\d+)\s+years', text.lower())
    return exp[0] if exp else "0"
```

```python
results = []

for file in os.listdir("resumes"):

    path = os.path.join("resumes", file)

    if file.endswith(".pdf"):
        text = extract_text_from_pdf(path)
    else:
        text = extract_text_from_docx(path)

    name = extract_name(text)
    email, phone = extract_contact_details(text)
    skills = extract_skills(text)
    experience = extract_experience(text)

    results.append({
        "Name": name,
        "Email": email,
        "Phone": phone,
        "Skills": ", ".join(skills),
        "Experience": experience
    })

df = pd.DataFrame(results)
df
```

```python
df.to_excel("Candidates.xlsx", index=False)
print("Excel created successfully")
```
